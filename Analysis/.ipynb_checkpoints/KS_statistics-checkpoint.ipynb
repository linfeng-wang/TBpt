```python
import pandas as pd
import numpy as np
from scipy import stats

# -----------------------------
# DeLong implementation
# -----------------------------
def compute_midrank(x):
    idx = np.argsort(x)
    sorted_x = x[idx]
    n = len(x)
    midranks = np.zeros(n, dtype=float)
    i = 0
    while i < n:
        j = i
        while j < n and sorted_x[j] == sorted_x[i]:
            j += 1
        midranks[i:j] = 0.5 * (i + j - 1)
        i = j
    out = np.empty(n, dtype=float)
    out[idx] = midranks + 1
    return out

def fastDeLong(predictions_sorted_transposed, label_1_count):
    m = label_1_count
    n = predictions_sorted_transposed.shape[1] - m
    pos = predictions_sorted_transposed[:, :m]
    neg = predictions_sorted_transposed[:, m:]
    k = predictions_sorted_transposed.shape[0]

    tx = np.array([compute_midrank(pred) for pred in pos])
    ty = np.array([compute_midrank(pred) for pred in neg])
    tz = np.array([compute_midrank(predictions_sorted_transposed[i]) for i in range(k)])

    aucs = tz[:, :m].sum(axis=1) / (m*n) - (m+1)/(2*n)
    v01 = (tx - tx.mean(axis=1)[:, None]) / n
    v10 = (ty - ty.mean(axis=1)[:, None]) / m
    sx = np.cov(v01)
    sy = np.cov(v10)
    s = sx/m + sy/n
    return aucs, s

def delong_auc_variance(y_true, y_scores):
    y_true = np.array(y_true)
    y_scores = np.clip(np.array(y_scores), 1e-7, 1 - 1e-7)

    order = np.argsort(-y_scores)
    y_scores_sorted = y_scores[order]
    y_true_sorted = y_true[order]

    label_1_count = int(y_true.sum())
    preds_sorted = np.expand_dims(y_scores_sorted, 0)

    aucs, s = fastDeLong(preds_sorted, label_1_count)

    # Handle scalar covariance case
    if np.isscalar(s):
        return aucs[0], float(s)
    else:
        return aucs[0], s[0, 0]

def delong_test(y_true, A, B):
    # A = np.clip(A, 1e-9, 1 - 1e-9)
    # B = np.clip(B, 1e-9, 1 - 1e-9)

    aucA, varA = delong_auc_variance(y_true, A)
    aucB, varB = delong_auc_variance(y_true, B)

    diff = aucA - aucB
    se = np.sqrt(varA + varB)
    z = diff / se
    p = 2 * (1 - stats.norm.cdf(abs(z)))

    return aucA, aucB, diff, z, p

# -----------------------------
# Load data
# -----------------------------
df = pd.read_csv("./probs_output.csv")

y = df["y_te2"].values
model_cols = [c for c in df.columns if c.startswith("y_prob")]

# Clip all models
model_data = {m: np.clip(df[m].values, 1e-7, 1 - 1e-7) for m in model_cols}

# -----------------------------
# Pairwise DeLong comparison
# -----------------------------
results = []

for i in range(len(model_cols)):
    for j in range(i + 1, len(model_cols)):
        m1 = model_cols[i]
        m2 = model_cols[j]

        aucA, aucB, diff, z, p = delong_test(y, model_data[m1], model_data[m2])

        results.append([m1, m2, aucA, aucB, diff, z, p])

results_df = pd.DataFrame(
    results,
    columns=["Model A", "Model B", "AUC A", "AUC B", "Difference", "Z", "p value"]
)

print(results_df)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ks_2samp

# Load your file

# Extract prediction columns
model_cols = [c for c in df.columns if c.startswith("y_prob")]

results = []

# Pairwise KS tests
for i in range(len(model_cols)):
    for j in range(i+1, len(model_cols)):
        m1 = model_cols[i]
        m2 = model_cols[j]

        # Perform 2-sample KS test
        stat, p_value = ks_2samp(df[m1].values, df[m2].values)

        results.append([m1, m2, stat, p_value])

# Convert to a dataframe for readability
ks_df = pd.DataFrame(results, columns=["Model A", "Model B", "KS Statistic", "p value"])
print(ks_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp

# Load your file

# Original model columns
model_cols = [c for c in df.columns if c.startswith("y_prob")]

# Replace with your chosen model names
model_names = [
    "Logistic Reg.",
    "Naive Bayes",
    "SVM",
    "k-NN",
    "MLP",
    "CatBoost",
    "LightGBM",
    "XGB"
]

n = len(model_cols)

# Compute KS matrix
ks_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        if i == j:
            ks_matrix[i, j] = 0.0
        else:
            stat, _ = ks_2samp(df[model_cols[i]], df[model_cols[j]])
            ks_matrix[i, j] = stat

# Plot heatmap
plt.figure(figsize=(9, 7))
plt.imshow(ks_matrix, cmap="Reds_r", interpolation="nearest")
plt.colorbar(label="KS Statistic")

# Axis labels
plt.xticks(range(n), model_names, rotation=45, ha="right")
plt.yticks(range(n), model_names)

# Write KS values on the grid
for i in range(n):
    for j in range(n):
        value = ks_matrix[i, j]
        plt.text(j, i, f"{value:.2f}", ha="center", va="center",
                 color="white" if value > 0.5 else "black")

plt.title("KS Statistic Heatmap Comparing Model Probability Distributions")
plt.tight_layout()
plt.show()
```
